```python
paternal = df[df['imprint_status'] == 'paternal']['length']
maternal = df[df['imprint_status'] == 'maternal']['length']

# Ð¢ÐµÑÑ‚ ÐœÐ°Ð½Ð½Ð°-Ð£Ð¸Ñ‚Ð½Ð¸ (Ð½ÐµÐ¿Ð°Ñ€Ð°Ð¼ÐµÑ‚Ñ€Ð¸Ñ‡ÐµÑÐºÐ¸Ð¹)
u_stat, p_value = stats.mannwhitneyu(paternal, maternal)
print(f"Ð¢ÐµÑÑ‚ ÐœÐ°Ð½Ð½Ð°-Ð£Ð¸Ñ‚Ð½Ð¸: U={u_stat}, p={p_value:.4f}")

# Ð’Ñ‹Ð²Ð¾Ð´ Ð¸Ð½Ñ‚ÐµÑ€Ð¿Ñ€ÐµÑ‚Ð°Ñ†Ð¸Ð¸
if p_value < 0.05:
    print("Ð¡ÑƒÑ‰ÐµÑÑ‚Ð²ÑƒÐµÑ‚ ÑÑ‚Ð°Ñ‚Ð¸ÑÑ‚Ð¸Ñ‡ÐµÑÐºÐ¸ Ð·Ð½Ð°Ñ‡Ð¸Ð¼Ð¾Ðµ Ñ€Ð°Ð·Ð»Ð¸Ñ‡Ð¸Ðµ Ð² Ð´Ð»Ð¸Ð½Ð°Ñ… Ñ€ÐµÐ³Ð¸Ð¾Ð½Ð¾Ð²")
else:
    print("ÐÐµÑ‚ Ð·Ð½Ð°Ñ‡Ð¸Ð¼Ð¾Ð³Ð¾ Ñ€Ð°Ð·Ð»Ð¸Ñ‡Ð¸Ñ Ð² Ð´Ð»Ð¸Ð½Ð°Ñ… Ñ€ÐµÐ³Ð¸Ð¾Ð½Ð¾Ð²")
```

```python
# Ð¡Ð¾Ð·Ð´Ð°ÐµÐ¼ Ñ‚Ð°Ð±Ð»Ð¸Ñ†Ñƒ ÑÐ¾Ð¿Ñ€ÑÐ¶ÐµÐ½Ð½Ð¾ÑÑ‚Ð¸
contingency_table = pd.crosstab(df['chrom'], df['imprint_status'])

# Ð¢ÐµÑÑ‚ Ñ…Ð¸-ÐºÐ²Ð°Ð´Ñ€Ð°Ñ‚
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)
print(f"\nÐ¢ÐµÑÑ‚ Ñ…Ð¸-ÐºÐ²Ð°Ð´Ñ€Ð°Ñ‚: Ï‡Â²={chi2:.2f}, p={p:.4f}")

# Ð’Ñ‹Ð²Ð¾Ð´ Ð¸Ð½Ñ‚ÐµÑ€Ð¿Ñ€ÐµÑ‚Ð°Ñ†Ð¸Ð¸
if p < 0.05:
    print("Ð Ð°ÑÐ¿Ñ€ÐµÐ´ÐµÐ»ÐµÐ½Ð¸Ðµ Ñ‚Ð¸Ð¿Ð¾Ð² Ð¸Ð¼Ð¿Ñ€Ð¸Ð½Ñ‚Ð¸Ð½Ð³Ð° Ð¿Ð¾ Ñ…Ñ€Ð¾Ð¼Ð¾ÑÐ¾Ð¼Ð°Ð¼ Ð½ÐµÑÐ»ÑƒÑ‡Ð°Ð¹Ð½Ð¾")
else:
    print("Ð Ð°ÑÐ¿Ñ€ÐµÐ´ÐµÐ»ÐµÐ½Ð¸Ðµ Ñ‚Ð¸Ð¿Ð¾Ð² Ð¸Ð¼Ð¿Ñ€Ð¸Ð½Ñ‚Ð¸Ð½Ð³Ð° Ð¿Ð¾ Ñ…Ñ€Ð¾Ð¼Ð¾ÑÐ¾Ð¼Ð°Ð¼ ÑÐ»ÑƒÑ‡Ð°Ð¹Ð½Ð¾")
```

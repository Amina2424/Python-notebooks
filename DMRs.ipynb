```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Загрузка данных (предполагается, что файл в кодировке UTF-16)
df = pd.read_csv('/content/метилирование.tsv', sep='\t')

# Переименование колонок для удобства
df.columns = ['chrom', 'start', 'end', 'imprint_status']

# Расчет длины региона
df['length'] = df['end'] - df['start']
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='chrom', hue='imprint_status', order=sorted(df['chrom'].unique()))
plt.title('Распределение импринтированных регионов по хромосомам')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='imprint_status', y='length', showfliers=False)
plt.title('Распределение длин регионов по типу импринтинга')
plt.ylabel('Длина региона (п.н.)')
plt.xlabel('Тип импринтинга')
plt.show()
```

```python
status_counts = df['imprint_status'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(status_counts, labels=status_counts.index, autopct='%1.1f%%')
plt.title('Соотношение типов импринтинга')
plt.show()
```

```python
paternal = df[df['imprint_status'] == 'paternal']['length']
maternal = df[df['imprint_status'] == 'maternal']['length']

# Тест Манна-Уитни (непараметрический)
u_stat, p_value = stats.mannwhitneyu(paternal, maternal)
print(f"Тест Манна-Уитни: U={u_stat}, p={p_value:.4f}")

# Вывод интерпретации
if p_value < 0.05:
    print("Существует статистически значимое различие в длинах регионов")
else:
    print("Нет значимого различия в длинах регионов")
```

```python
# Создаем таблицу сопряженности
contingency_table = pd.crosstab(df['chrom'], df['imprint_status'])

# Тест хи-квадрат
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)
print(f"\nТест хи-квадрат: χ²={chi2:.2f}, p={p:.4f}")

# Вывод интерпретации
if p < 0.05:
    print("Распределение типов импринтинга по хромосомам неслучайно")
else:
    print("Распределение типов импринтинга по хромосомам случайно")
```

```python
top10 = df.nlargest(10, 'length')[['chrom', 'start', 'end', 'length', 'imprint_status']]
print("\nТоп-10 самых длинных регионов:")
print(top10.to_string(index=False))
```

```python
plt.figure(figsize=(14, 6))
chrom_order = sorted(df['chrom'].unique(), key=lambda x: int(x[3:]) if x[3:].isdigit() else 25)
ax = sns.countplot(data=df, x='chrom', order=chrom_order, palette='viridis')
plt.title('Общее количество ICR по хромосомам', pad=20)
plt.xlabel('Хромосома')
plt.ylabel('Количество регионов')

# Добавляем аннотации с количеством
for p in ax.patches:
    ax.annotate(f'{p.get_height()}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center',
                xytext=(0, 5),
                textcoords='offset points')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))

# Гистограмма с кривой плотности
sns.histplot(df['length'], bins=50, kde=True, color='royalblue')
plt.title('Распределение длин всех потенциальных ICR', pad=15)
plt.xlabel('Длина региона (п.н.)')
plt.ylabel('Частота')
plt.axvline(df['length'].median(), color='red', linestyle='--',
            label=f'Медиана: {df["length"].median():.0f} п.н.')
plt.legend()

# Вставка с zoom на короткие регионы
ax_inset = plt.axes([0.5, 0.5, 0.35, 0.35])
sns.histplot(df[df['length'] <= 500]['length'], bins=30, ax=ax_inset, color='green')
ax_inset.set_title('Регионы ≤500 п.н.')
ax_inset.set_xlabel('')
ax_inset.set_ylabel('')

plt.tight_layout()
plt.show()
```

```python
chrom_cumulative = df.groupby('chrom')['length'].sum().sort_values(ascending=False)
chrom_cumulative_percent = chrom_cumulative.cumsum() / chrom_cumulative.sum() * 100

plt.figure(figsize=(10, 6))
chrom_cumulative_percent.plot(kind='bar', color='purple')
plt.title('Кумулятивный процент покрытия ДНК импринтированными регионами', pad=15)
plt.ylabel('Кумулятивный % от общего покрытия')
plt.xlabel('Хромосома')
plt.axhline(80, color='red', linestyle='--', alpha=0.5)
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x='start', y='length', hue='chrom',
                palette='tab20', alpha=0.7, s=50)
plt.title('Зависимость длины региона от его геномной позиции', pad=15)
plt.xlabel('Стартовая позиция (п.н.)')
plt.ylabel('Длина региона (п.н.)')
plt.yscale('log')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
shapiro_test = stats.shapiro(df['length'])
print(f"Тест Шапиро-Уилка на нормальность:\nW={shapiro_test[0]:.4f}, p={shapiro_test[1]:.4e}")

if shapiro_test[1] < 0.05:
    print("Распределение длин НЕ является нормальным (p < 0.05)")
else:
    print("Распределение длин соответствует нормальному (p >= 0.05)")
```
